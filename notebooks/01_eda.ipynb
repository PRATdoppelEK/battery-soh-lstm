# Battery SOH — Exploratory Data Analysis

This notebook explores the battery cycle dataset before modelling.
We examine SOH degradation curves, feature distributions, and correlations.

```python
import sys
sys.path.append('../src')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from data_loader import load_all_batteries

df = load_all_batteries()
print(df.shape)
df.head()
```

```python
# SOH degradation per battery
fig, ax = plt.subplots(figsize=(10, 4))
for battery_id, group in df.groupby('battery_id'):
    ax.plot(group['cycle_number'], group['soh'], label=battery_id, linewidth=1.5)
ax.axhline(0.8, color='red', linestyle='--', label='EOL threshold')
ax.set_xlabel('Cycle number')
ax.set_ylabel('State of Health (SOH)')
ax.set_title('Battery SOH Degradation')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Feature correlation heatmap
import matplotlib.pyplot as plt
import numpy as np

feature_cols = ['discharge_capacity','voltage_mean','voltage_min',
                'current_mean','temperature_mean','temperature_max',
                'discharge_time','soh']
corr = df[feature_cols].corr()

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(corr, cmap='RdBu_r', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax)
ax.set_xticks(range(len(feature_cols)))
ax.set_yticks(range(len(feature_cols)))
ax.set_xticklabels(feature_cols, rotation=45, ha='right', fontsize=9)
ax.set_yticklabels(feature_cols, fontsize=9)
ax.set_title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
