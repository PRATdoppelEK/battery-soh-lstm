```python
import matplotlib.pyplot as plt
import numpy as np

# SOH distribution in each split
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
for ax, split, color in zip(axes,
    ['train','val','test'], ['#2563EB','#059669','#DC2626']):
    y = scaled[f'y_{split}']
    ax.hist(y, bins=30, color=color, alpha=0.8)
    ax.set_title(f'{split.capitalize()} SOH distribution')
    ax.set_xlabel('SOH')
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```
